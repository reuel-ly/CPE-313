# Data Extraction

```python
import pandas as pd

data = "/content/drive/MyDrive/Colab Notebooks/Dataset/twitter_training.csv"
data = pd.read_csv(data)
data.head()
```

```python
pd.set_option('display.max_colwidth', None)
```

# 1. Find a specific word on the twitter content and number of occurrences

```python
# Check for 'hello' word
import re

word_count = data.iloc[:, 3].apply(lambda x : len(re.findall('hello', str(x))))
word_count
```

```python
word_count.sum()
```

```python
data.iloc[word_count.argmax()]
```

> The total amount of the word 'hello' in the data is 79. I checked one instance using the argmax() function.

# 2. Find  a specific special character and number of occurrences

```python
# Check for '!' special character

word_count = data.iloc[:, 3].apply(lambda x : len(re.findall('!', str(x))))
word_count
```

```python
word_count.sum()
```

```python
data.iloc[word_count.argmax()]
```

# 3. Find ALL numeric characters  and number of occurrences


```python
# Check for '!' special character

word_count = data.iloc[:, 3].apply(lambda x : len(re.findall('[0-9]', str(x))))
word_count
```

```python
word_count.sum()
```

```python
data.iloc[word_count.argmax()]
```

> For this part, I simply used [0-9] to just get all the instance of numbers. In each datapoints.

# 4. Find ALL URLs and remove the URLs.

```python
# Check for URLs special character
filter = r"https?://[^\s]+|www\.[^\s]+|bit\.ly/[^\s]+|ttps?://[^\s]+"
word_count = data.iloc[:, 3].apply(lambda x : len(re.findall(filter,
                                                             str(x))))
word_count
```

```python
word_count.sum()
```

```python
idx_with_url = word_count.argmax()
```

```python
data.iloc[idx_with_url]
```

```python
filter = r"https?://[^\s]+|www\.[^\s]+|bit\.ly/[^\s]+|ttps?://[^\s]+"
clean_data = data.iloc[:, 3].apply(lambda x : re.sub(filter, "",
                                                     str(x)))
clean_data
```

```python
clean_data.iloc[idx_with_url]
```

> For this part, instead of just detecting https://, I added bit.ly because at my first execution. The bit.ly and other links are not deleted.
