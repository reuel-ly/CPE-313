Find the lyrics by the artist The Beatles

```python
beatles = data[data.artist=='The Beatles']
beatles.head()
```

Remove characters from the lyrics

```python
import re

newline = lambda x: re.sub(r'\n', ' ', x) # remove \n
alphanumeric = lambda x: re.sub(r'\w*\d\w*', ' ', x) # remove alphanumeric words

corpus = beatles.text.map(newline).map(alphanumeric)
corpus.head()
```

Create a document term matrix using Count Vectorizer with the stop words turned on to English

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words="english")
X = cv.fit_transform(corpus).toarray()

dt = pd.DataFrame(X, columns=cv.get_feature_names_out()).set_index(beatles.song)
dt.head()
```

Set display options so that the lyrics aren't cut off

```python
pd.set_option('display.max_colwidth', None)
```

```python
beatles[beatles.song=='Imagine'].text
```

Imagine lyrics in Count Vectorizer form


```python
imagine = list(dt.loc['Imagine'])
imagine[:20]
```

Define the cosine similarity calculation

```python
from numpy import dot
from numpy.linalg import norm

cosine = lambda v1, v2: dot(v1, v2) / (norm(v1) * norm(v2))
```

Calculate all similarities and sort by the most similar

```python
similarity = [cosine(imagine, song) for song in X]
sorted(list(zip(similarity, beatles.song)), reverse=True)[1:6]
```

Create the TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(beatles.song)
dt_tfidf.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(beatles.song)
dt_tfidf.head()
```

Calculate all similarities and sort by the most similar using the TF-IDF data

```python
similarity_tfidf = [cosine(imagine, song) for song in X_tfidf]
sorted(list(zip(similarity_tfidf, beatles.song)), reverse=True)[1:6]
```

### **Discussion 5.1 | NLP Preprocessing**
* Tokenize (words)

* Tokenize (words with remove punctuation and all lowercase)
* Tokenize (words with remove punctuation, all lowercase and  remove stop words)

```python
import nltk
import re
import string

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
#@title Tokenize(words)
from nltk.tokenize import word_tokenize

tokens = data['OriginalTweet'].apply(word_tokenize)
print(tokens)
```

```python
#@title Tokenize (words with remove punctuation) / Tokenize (words with remove special characters and numbers)

clean_tokens = [word for sentence in tokens for word in sentence
                if word.isalpha()]
clean_tokens
```

```python
#@title Tokenize (words with remove punctuation and all lowercase)
clean_tokens = [word.lower() for word in clean_tokens if word.isalpha()]
clean_tokens
```

```python
#@title Tokenize (words with remove punctuation, all lowercase and remove stop words)
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

clean_tokens = [
    word.lower()
    for word in clean_tokens
    if word.isalpha() and word.lower() not in stop_words
]

clean_tokens
```

```python
# @title Tokenize (words with remove punctuation, all lowercase and remove stop words, with speech_tagging)
from nltk import pos_tag

clean_tokens = [
    word.lower()
    for word in clean_tokens
    if word.isalpha() and word.lower() not in stop_words
]

pos_tags = pos_tag(clean_tokens)

print(pos_tags)
```

### Data Cleaning for Text Similarity

```python
# dropping na rows

data.dropna(subset=['OriginalTweet'], inplace=True)
```

```python
# lowercasing

data.OriginalTweet = data.OriginalTweet.str.lower()
```

```python
import re

def remove_noise(text):
    text = re.sub(r'http\S+', '', text)        # remove URLs
    text = re.sub(r'[^a-zA-Z\s]', '', text)   # remove numbers & special chars
    text = re.sub(r'\s+', ' ', text).strip()  # remove extra whitespace
    return text

data.OriginalTweet = data.OriginalTweet.apply(remove_noise)
```

```python
# Tokenize
data['tokens'] = data.OriginalTweet.apply(word_tokenize)
```
